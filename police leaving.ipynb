```python
def new_hire(df):
    count = 0
    for i in range(len(df.columns)):
        
        if i == len(df.columns) - 1:
            return(count)
        for j in range(len(df)):           
            id_num = df.iloc[j][i]
            id_num2 = df.iloc[j][i+1]
            if np.isnan(id_num2):
                count += 0
            elif id_num == id_num2:
                count += 0
            else:
                count += 1
        
        
        
```

```python
def leavers(df):
    count = 0
    for i in range(len(df.columns)):
        if i == len(df.columns) - 1:
            return(count)
        for j in range(len(df)):           
            id_num = df.iloc[j][i]
            id_num2 = df.iloc[j][i+1]
            if np.isnan(id_num2):
                if np.isnan(id_num):
                    count += 0
                else:
                    count += 1
            elif id_num == id_num2:
                count += 0
            elif np.isnan(id_num):
                count += 0
            else:
                count += 1
        
```

```python
def avg_employee(df):
    month = []
    for i in range(len(df.columns)):
        count = 0
        for j in range(len(df)):
            if type(df.iloc[j][i]) == int:
                count += 1
        month.append(count)
    return(np.mean(month))
```
